Developers notes
================

The following pages are helpful:

- See here for github pages workflow: https://daler.github.io/sphinxdoc-test/includeme.html
 - TL;DR: run `$ make html` in PyPlr/docs, then go to PyPlr-sphinxdocs/html, check you are on the gh-pages branch (`$ git branch`), and run `$ git commit -a -m "rebuilt docs"` followed by `git push origin gh-pages`
- `python setup.py develop` makes recent changes instantly available to sys.path. When finished developing, `python setup.py develop --uninstall` removes from easy-install.pth and deletes the .egg-link. This avoids having to say `sys.path.insert(0, '../')` in notebooks.
- See here for how to package python projects: https://packaging.python.org/tutorials/packaging-projects/
- Current workflow for new release is `$ python setup.py sdist` followed by `$ python -m twine upload dist/*`
