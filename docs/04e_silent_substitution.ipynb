Silent substitution
===================

The human retina has five classes of photoreceptor: rods, cones (S, M, L) and intrinsically photosenstive retinal ganglion cells (ipRGC). Rods are most numerous (77-107 million cells) and are responsible for scotopic (low light) vision, whereas cones (4-5 million cells) are responsible for mesopic / photic vision. The ipRGCs are a special subpopulation of ganglion cell which express the photopigment melanopsin in their axons and soma, making them intrinsically photosensetive. They do not mediate vision like rods and cones, but they are known to regulate steady state pupil size. Each of the photoreceptors are tuned to different wavelengths of light, as described by the photopigment spectral sensetivity curves, which effectively describes the probability that a photoreceptor will absorb a photon of light at a given wavelength.
