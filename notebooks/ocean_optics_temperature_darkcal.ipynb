```python
board_temps = data.board_temperature.unique()
integration_times = data.integration_time.unique()

cols = ['{} deg. C'.format(col) for col in board_temps]
rows = ['Micros:\n{}'.format(row) for row in integration_times]

fig, axes = plt.subplots(nrows=8, ncols=6, sharex=True, sharey=True, figsize=(16, 16))

for row, it in zip(axes, integration_times):
    for ax, bt in zip(row, board_temps):
        d = data.loc[((data.board_temperature==bt)&(data.integration_time==it))]
        ax.plot(d.wavelengths, d.raw, lw=.5, color='r')
    
    
for ax, col in zip(axes[0], cols):
    ax.set_title(col)

for ax, row in zip(axes[:,0], rows):
    ax.set_ylabel(row, size='large')

fig.tight_layout()
plt.show()
fig.savefig('ocean_optics_temperature_darkcal.png')
```
